# Matplotlib
It is a plotting library for python 

```python
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2d')
plt.savefig('Test.png')
```

```python
z=np.array(x*x)
v=(z*2)+1.345


```

```python
plt.plot(x,v,'b')
```

```python
plt.plot(x,v,'r*',linestyle='dashed',linewidth=2,markersize=12)
```

## Creating Subplots:


```python
plt.subplot(2,2,1) # plt.subplot(rows,columns,position of the following plot)
plt.plot(x,v,'b')
plt.subplot(2,2,2)
plt.plot(x,v,'r*',markersize=14)
plt.subplot(2,2,3)
plt.plot(x,v,'go--')
```

```python
# compute the sine wave
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,'g')
plt.title('Sine Wave Form')
```

```python
# Bar plot
x=[3,4,5]
y=[10,16,21]
plt.bar(x,y)
# can also put another value( bar graph) to compare
```

```python
# Pie chart
labels='python','java','c++','ruby'
sizes=[215,245,210,130]
colors=['gold','green','coral','blue']
explode=(.4,0,0,0) # indicates that the first mentioned label should be exploded upon viewing
plt.pie(sizes,explode=explode,labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)
plt.show()
        
```
